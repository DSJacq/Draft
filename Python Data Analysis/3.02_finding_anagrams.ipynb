```python
wordclear = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordclear[:10]
```

```python
sorted('lives')
```

```python
sorted('haye') == sorted('love')
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
signature('live')
```

```python
'/'.join(['a','b','c'])
```

```python
def anagram(myword):
    return [word for word in wordclear if signature(word) == signature(myword)]
```

```python
anagram('dictionary')
```

```python
%timeit anagram('dictionary')
```

```python
import collections
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(word):
    return words_bysig[signature(word)]
```

```python
anagram_fast('dictionary')
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
len(anagrams_all)
```
